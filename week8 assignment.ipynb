```python
%pylab inline

import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
# 1. Load the data in the (yellow highlighted) Data
# Folder in the file auto-mpg.data into a
# pandas DataFrame.

# 3.Use the attribute information provided in
# auto-mpg.names to name your pandas
# DataFrame columns.

df = pd.read_fwf('auto-mpg.data', header=None, names=['displacement', 'mpg', 'cylinders', 'horsepower',
                                                     'weight','acceleration','model_year','origin','car_name'])
df.head()
```

# Review the data and the provided data set description.

### This data shows the car models and its technical infromation such as horsepower and speed, cylinder and weight.

```python
# The origin column values are 1, 2, and 3.
# Convert these values respectively to ‘USA’,
# ‘Asia’, and ‘Europe
df['origin'] = df['origin'].replace({1: 'USA', 2: 'Asia', 3: 'Europe'})
df
```

```python
# Create a bar chart that shows the distribution for cylinders.
df['cylinders'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
plt.show()
```

```python
# Create a scatterplot that shows the relationship between horsepower and weight.
df.plot.scatter(x='horsepower', y='weight')
plt.title('Relationship between Horsepower and Weight')
```
